```python
import pandas as pd
import scipy.stats as st
import seaborn as sns

data= pd.read_csv("Heart_Disease_Prediction.csv")
data.info()
data
```

```python
#Plot scatter Plot
data.plot(kind='scatter',x="Age",y="BP")
```

```python
list(data.columns)
```

```python
#to ctreate line of best fit
sns.lmplot(x="Age", y="BP",data=data);
```

```python
#to calculate pearson r coefficient
st.pearsonr(data['BP'], data['Cholesterol'])
```

```python
#create correlation matrix
#A correlation matrix is a handy way to calculate the pairwise correlation coefficients between two or more (numeric) variables
cormat = data.corr()
round(cormat,2)
```

```python
#create heatmap
sns.heatmap(cormat);
```

```python
data.plot(kind='scatter',x="Age",y="ST depression")
```

```python
data.plot(kind='scatter',x="Sex",y="Cholesterol")
sns.lmplot(x="Sex", y="Cholesterol",data=data);
```

```python
#Show row percentages
conttab_relfreq = pd.crosstab(
    data['Age'],
    data['Sex'],
    margins = True,
    normalize='index'
   )
conttab_relfreq
```

```python
#Chi square test of independence
contab_obs = pd.crosstab(
    data['Age'],
    data['BP'],
    margins = False)
chi = st.chi2_contingency(contab_obs)
chi
```
